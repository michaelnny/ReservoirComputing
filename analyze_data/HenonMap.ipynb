```python
import numpy as np
import matplotlib.pyplot as plt

def henon_map(x, y, a=1.4, b=0.3):
    new_x = 1 - a * x**2 + y
    new_y = b * x
    return new_x, new_y

# Number of iterations
iterations = 10000

# Initial values
x_values = [0]
y_values = [0]

# Generate Henon map data
for i in range(iterations):
    x, y = henon_map(x_values[-1], y_values[-1])
    x_values.append(x)
    y_values.append(y)

# Plot the Henon map
plt.figure(figsize=(8, 6))
plt.scatter(x_values, y_values, s=1, c='blue', marker='.')
plt.title('Henon Map')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
